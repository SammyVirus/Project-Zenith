```python
import os
import numpy as np
import cv2
import tqdm
import matplotlib.pyplot as plt
from sklearn.utils import shuffle
```

```python
size = 240
x = []
y = []
```

```python
dir_box = os.listdir('boxes')
#dir_box.remove('.DS_Store')

for i in tqdm.tqdm(dir_box):
    img = cv2.imread('boxes/' + str(i))
    img = cv2.resize(img, (size, size),  interpolation = cv2.INTER_AREA)
    x.append(img)
    y.append(1)
```

```python
len(x)
```

```python
dir_box = os.listdir('no_boxes')
#dir_box.remove('.DS_Store')

for i in tqdm.tqdm(dir_box[:1200]):
    img = cv2.imread('no_boxes/' + str(i))
    img = cv2.resize(img, (size, size),  interpolation = cv2.INTER_AREA)
    x.append(img)
    y.append(0)
```

```python
x = np.array(x)
y = np.array(y)
```

```python
x.shape
```

```python
for i in tqdm.tqdm(range(x.shape[0])):
    #X[i] = cv2.GaussianBlur(X[i], (3,3), 0)
    x[i] = cv2.fastNlMeansDenoisingColored(x[i],None,10,10,7,21)
```

```python
X, y = shuffle(x, y)
```

```python
X = X/255
```

```python
np.save('train_data', X)
np.save('train_label', y)
```
